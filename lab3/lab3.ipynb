## 10. Jakie kategorie filmów zostały wypożyczone przez klientów?

```python
df = pd.read_sql_query("select distinct(category.name) from ((((category INNER JOIN film_category ON category.category_id = film_category.category_id) INNER JOIN film ON film_category.film_id = film.film_id) INNER JOIN inventory ON film.film_id = inventory.film_id) INNER JOIN rental ON inventory.inventory_id = rental.inventory_id) where rental_date between '1900-01-01 00:00:00' and '2111-02-01 23:59:59'",con=connection)
df
```
